```python
#selecting the column from the dataset
fulldata['Cut Off Merit (Score)']=fulldata['Cut Off Merit (Score)'].astype('str')
```

```python
#creating the fucntion for the splitting the score 

def score_f(x):
    x=x.split('(')
    x=x[1]
    x=x.split(')')
    x=x[0]
    x=int(x)
    return x
```

```python
score_data=score_data.dropna()
```

```python
score_data=score_data.apply(score_f)
```

```python
#crearting a new column for the score 

fulldata['score']=score_data
```

```python
x=fulldata[['Exam (JEE/MHT- CET)', 'Type', 'Seat Type', 'score']]
from sklearn.preprocessing import LabelEncoder
for i in range(3):
    x.iloc[:,i]=LabelEncoder().fit_transform(x.iloc[:,i])
```

```python
y=fulldata['Course Name']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.tree import DecisionTreeClassifier ,plot_tree
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
mode_tree=DecisionTreeClassifier().fit(x_train,y_train)
mode_log=LogisticRegression().fit(x_train,y_train)
pred_tree=mode_tree.predict(x_test)
pred_log=mode_log.predict(x_test)
```

```python
print(confusion_matrix(pred_log,y_test))
accuracy_score(pred_log,y_test)*100
```

```python
print(confusion_matrix(pred_tree,y_test))
accuracy_score(pred_tree,y_test)*100
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(30,20))
plot_tree(mode_tree,filled=True)
plt.show()
```
