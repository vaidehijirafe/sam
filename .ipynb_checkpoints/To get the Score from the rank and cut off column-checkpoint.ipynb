```python
#selecting the column from the dataset
fulldata['Cut Off Merit (Score)']=fulldata['Cut Off Merit (Score)'].astype('str')
```

```python
score_data=fulldata['Cut Off Merit (Score)']
```

```python
#creating the fucntion for the splitting the score 

def score_f(x):
    x=x.split('(')
    x=x[1]
    x=x.split(')')
    x=x[0]
    x=int(x)
    return x
```

```python
score_data=score_data.dropna()
```

```python
#applying the function to the score_data
#score_data=score_data.apply(lambda x: score_f(x))
x={}
for i in range(len(score_data)):
    x[i]=score_f(score_data[i])
```

```python
for i in range(len(x)):
    fulldata['score'][i]=x[i]
    
    
```

```python
#crearting a new column for the score 

fulldata['score']=x
```

```python
fulldata.to_csv('dataset1.csv')
```
