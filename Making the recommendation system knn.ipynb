```python
#ploting the average rating vs the course name
score_val=pd.DataFrame(dataset6.groupby('Course Name')['score'].mean())
score_val.sort_values('score',ascending=False).head(50)
```

```python
dataset6.groupby('Course Name')['score'].count().sort_values(ascending=False).head()
```

```python
score_val=score_val.reset_index()
```

```python
sam=dataset6.groupby('Course Name')['score'].count()
```

```python
#creating a new column where we have to show number of times the score has appered with respect to coursename
for i in range(len(sam)):
    score_val.loc[i,'num_score']=sam[i]
    print(score_val.loc[i,'num_score'])
```

```python
#ploting the number of times the score has appeared
plt.figure(figsize=(10,4))
score_val['num_score'].hist(bins=70)
```

```python
#ploting the avg(mean) of the score has appeared
plt.figure(figsize=(10,4))
score_val['score'].hist(bins=70)
```

```python
#creating the joint plot 

sns.jointplot(x='score',y='num_score',data=score_val,alpha=0.5)
```
